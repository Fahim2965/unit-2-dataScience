# Streaming Viewers Data Analysis 

Streaming membership apps have gained widespread popularity, attracting a diverse clientele who prefer the flexibility of watching their favorite shows at their own convenience, a feature notably absent in traditional cable subscriptions. The cost-effectiveness of streaming services, coupled with the absence of ads that often accompany cable television, makes them an exelent choice for many users. Moreover, these apps offer a variety of membership tiers, allowing users to customize their plans based on individual preferences and needs.

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('https://github.com/dhaitz/matplotlib-stylesheets/raw/master/pitayasmoothie-dark.mplstyle')
df = pd.read_csv('streaming_viewership_data (1).csv')
df.head()
```

# 1. Genre's watched by each gender

```python
dos = df.groupby(['Gender', 'Genre'])['Interaction_Events'].sum()
dos.plot(kind='bar', color=('#F98870','#F970E0'))
plt.title('People per genre by gender')
plt.ylabel('Total amount of people')
plt.xlabel('Gender and Gender')
```

The graph illustrates that males and females have different preferences in terms of the genres they enjoy watching. Notably, females show a particular interest in Thrillers as their most-watched genre, while males tend to favor Documentaries as their top choice for viewing. This discrepancy highlights the diverse tastes between genders when it comes to preferred genres.

# 2. Most watched genre in most country's

```python
dfs = pd.read_csv('split.csv')

art = dfs.groupby(['Country', 'Genre'])['Interaction_Events'].mean()
art.plot(kind= 'bar', color=('#70F981','#8B3000'))
```

The graph illustrates the popularity of different genres across various countries. It implies that people residing in the same country share similar interests, potentially influenced by factors such as religion and culture. This observation suggests a connection between entertainment preferences and the cultural and societal characteristics of a particular region.

# 3. Highest rated Genres

```python
at = df.groupby(['Genre'])['Ratings'].sum()
at.plot(kind= 'bar', color=('#FFFF68','#8B0556'))
```

The graph highlights the most popular genres within the dataset, providing valuable information for potential advertising strategies, especially for new streaming apps. By identifying the preferred genres, advertisers can tailor their promotions to target audiences more effectively. Using this data could enhance the marketing efforts of streaming platforms, aligning content offerings with the preferences of their potential users and increasing the chances of attracting a larger audience.

# 4. Most popular Languages for each Genre

```python
bts = df.groupby(['Languages','Genre'])['Interaction_Events'].sum()
bts.plot(kind= 'bar', color=('#FFFF48','#8B0686'))
```

The graph depicts the predominant languages in which each genre is viewed or watched. It provides insights into the linguistic preferences associated with specific genres. Understanding the language patterns associated with popular genres can be beneficial for content creators and distributors, allowing them to optimize language options and cater to the linguistic diversity of their audience.
